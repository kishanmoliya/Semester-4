### 01) WAP to demonstrate the use of Pie chart.

```python
x = [10, 20, 30, 40]
y = ["Food", "Travel", "Running", "shopping"]
c = ['r', 'g', 'y', 'b']
ex = [0.5, 0.0, 0.0, 0.0]
plt.pie(x,labels=y, colors=c, explode=ex, shadow=True, radius=1.2, autopct="%0.2f%%", labeldistance=1.2)
plt.title("expensics")
plt.show()
```

### 02) WAP to to Plot List random of X, Y Coordinates in Matplotlib.

```python
import random as r
x = [10, 20, 30, 40]
y = [r.randint(0, 101) for i in range(4)]

plt.plot(x, y)
plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
x = ['python', 'c', 'java', 'c++']
y = [80, 60, 65, 50]
c = ['r', 'g', 'y', 'b']
plt.bar(x, y, color=c, edgecolor='b', alpha=0.8)
plt.title("Popularity of Language")
plt.xlabel('Language')
plt.ylabel('Popularity')
plt.show()
```

```python
x = ['python', 'c', 'java', 'c++']
y = [80, 60, 65, 50]
c = ['r', 'g', 'y', 'b']
plt.barh(x, y, color=c, edgecolor='b', alpha=0.8)
plt.title("Popularity of Language")
plt.xlabel('PopularityLanguage')
plt.ylabel('Language')
plt.show()
```

### 04) WAP to demonstrate the use of Histogram.

```python
import numpy as np
cpi = np.random.randint(0,10,100)
plt.hist(cpi,color='g')
plt.xlabel('Student')
plt.ylabel('SPI')
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x = ['python', 'c', 'java', 'c++']
y = [80, 60, 65, 50]
c = ['r', 'g', 'y', 'b']
plt.bar(x, y, color=c, edgecolor='b', alpha=0.8)
plt.title("Popularity of Language")
plt.xlabel('Language')
plt.ylabel('Popularity')

def addLabels(x, y):
    for i in range(len(x)):
        plt.text(i,y[i]+1,y[i])
addLabels(x,y)
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
c = ['r', 'g', 'y', 'b','k']
size = [100,200,300,400,500]
plt.scatter(x, y, color=c, marker='*', s=size,)
plt.title("Scatter plote")
plt.xlabel('Size')
plt.ylabel('Stars')
# plt.show()
plt.savefig('mychart.png')
```
